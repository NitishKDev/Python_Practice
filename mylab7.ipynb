Write a Python program that creates a m * n integer array and Prints its attributes using matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#creating array of m*n

m=int(input("Enter m value for row"))
n=int(input("Enter n value for column"))

#arr=np.arange(1,10).reshape(m,n) ..show error if m=3 n=4
arr=np.random.randint(1,10,size=(m,n))
print("Array:")
print(arr)
```

```python
# Print the attributes of the array
print("Array Attributes:")
print("Shape:", arr.shape)
print("Dimensions:", arr.ndim)
print("Data Type:", arr.dtype)
print("Length of each element of array in bytes is ", arr.itemsize)
```

```python
#visualize the array using matplotlib
plt.imshow(arr)
plt.colorbar(orientation='vertical')
plt.title("Integer array")
```
