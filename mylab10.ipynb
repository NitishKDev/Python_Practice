Write a Python program to demonstrate Time series analysis with Pandas.

```python
import pandas as pd
import numpy as np

# Generate some random time series data
dates = pd.date_range(start='2024-01-01', periods=100)
data = np.random.randn(100)

# Create a DataFrame
df = pd.DataFrame(data, index=dates, columns=['Value'])

# Display the DataFrame and basic analysis
print("Sample Data:")
print(df.head())
print("\nTime Series Analysis:")
print("Mean Value:", df['Value'].mean())
print("Standard Deviation:", df['Value'].std())
print("Maximum Value:", df['Value'].max())
print("Minimum Value:", df['Value'].min())

# Plot the time series data
df.plot(title='Time Series Data')
```
