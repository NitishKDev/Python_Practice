Implement a python program to demonstrate 
 Data visualization with various Types of Graphs using Numpy

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#pie chart
power=np.array([232,111,94,43])
name=["luffy","zoro","sanji","ussop"]
plt.pie(power,labels=name)
plt.show()
```

```python
#bar chart
power=np.array([232,111,94,43])
name=["luffy","zoro","sanji","ussop"]
plt.bar(name,power,color='y')
plt.title("Power chart")
plt.xlabel("Names")
plt.ylabel("Number")
plt.show()
```

```python
#line plot
x = np.linspace(0, 10, 100)
y=np.sin(x)
plt.plot(x,y,color='r')
plt.title("Line plot(sine wave)")
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.show()
```

```python
#scatter plot
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])

plt.scatter(x, y)
plt.show()
```

```python
#histogram
x = np.random.normal(170, 10, 120)
#x will be an array containing 120 random numbers that are normally distributed 
#around the mean value of 170,with a standard deviation of 10.
plt.hist(x)
plt.show() 
```
